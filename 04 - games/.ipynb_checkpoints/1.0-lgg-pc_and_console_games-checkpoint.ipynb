#### –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ `user_score`

–í –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –µ—Å—Ç—å 6701 –ø—Ä–æ–ø—É–≤—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ `user_score`. –û—Ü–µ–Ω–∫–∏ –∏–≥—Ä–æ–∫–æ–≤, –∫–∞–∫ –∏ –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤, —Å–∫–ª–∞–¥—ã–≤–∞—é—Ç—Å—è –∏–∑-–∑–∞ –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã—Ö –≤–ø–µ—á–∞—Ç–ª–µ–Ω–∏–π –æ–± –∏–≥—Ä–µ, –∏ –æ–Ω–∏ –º–æ–≥—É—Ç –±—ã—Ç—å –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã –æ—Ç –∂–∞–Ω—Ä–∞, –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –≥–æ–¥–∞ –≤—ã—Ö–æ–¥–∞.  –í–æ–∑–º–æ–∂–Ω–æ –¥–ª—è –≤—ã—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –ø–æ—Ä–æ–≥ –æ—Ü–µ–Ω–æ–∫ (–Ω–∞–ø—Ä–∏–º–µ—Ä, —á—Ç–æ–±—ã –∏–≥—Ä—É –æ—Ü–µ–Ω–∏–ª–∏ –±–æ–ª–µ–µ 100 –∏–≥—Ä–æ–∫–æ–≤), –∫–æ—Ç–æ—Ä—ã–π –Ω–µ –±—ã–ª –ø–µ—Ä–µ–π–¥–µ–Ω, –ø–æ—ç—Ç–æ–º—É –ø–æ—è–≤–∏–¥–∏—Å—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –¥–∞–Ω–Ω—ã—Ö, –∏–ª–∏ –µ—Å—Ç—å –æ—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–≥—Ä—É–∑–∫–µ –¥–∞–Ω–Ω—ã—Ö. –¢.–∫. –ø–æ–¥–æ–±–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –æ—á–µ–Ω—å –º–Ω–æ–≥–æ, —É–¥–∞–ª—è—Ç—å –∏—Ö –º—ã –Ω–µ –±—É–¥–µ–º, –∞ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –ø—Ä–∏—Å–≤–æ–µ–º –µ–º—É –∑–Ω–∞—á–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ —Ç–æ—á–Ω–æ –Ω–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è. –£ –Ω–∞—Å –æ—Ü–µ–Ω–∫–∞ —Ä–µ–π—Ç–∏–Ω–≥–∞ –æ—Ç 0 –¥–æ 10, –ø–æ—ç—Ç–æ–º—É –ø—Ä–∏—Å–≤–æ–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ 42.

```python
games['user_score'] = games['user_score'].fillna(value=42) 
```

```python
games['user_score'].isna().sum()
```

 - ¬´RP 17+¬ª (¬´Rating Pending¬ª) ‚Äî ¬´–†–µ–π—Ç–∏–Ω–≥ –æ–∂–∏–¥–∞–µ—Ç—Å—è - –ø—Ä–∏–º–µ—Ä–Ω–æ 17+¬ª: –ü—Ä–æ–¥—É–∫—Ç –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω ESRB –Ω–∞ –æ—Ü–µ–Ω–∫—É, –Ω–æ –ø—Ä–∏–º–µ—Ä–Ω–æ–µ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –±–ª–∏–∑–∫–æ –∫ —Ä–µ–π—Ç–∏–Ω–≥—É ¬´M¬ª. –î–∞–Ω–Ω–∞—è –º–∞—Ä–∫–∏—Ä–æ–≤–∫–∞ –±—ã–ª–∞ –≤–≤–µ–¥–µ–Ω–∞ —Å 2021 –≥. –∏ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—è–µ—Ç—Å—è –Ω–∞ –¥–µ–º–æ –∏–ª–∏ –ø—Ä–µ–∑–µ–Ω—Ç–∞—Ü–∏–æ–Ω–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç.  

–ö–∞–∫ –º—ã –≤–∏–¥–∏–º, –æ—Ç–¥–µ–ª—å–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–ª—è –∏–≥—Ä –±–µ–∑ —Ä–µ–π—Ç–∏–Ω–≥–∞ –Ω–µ—Ç, –ø–æ—ç—Ç–æ–º—É –ø—Ä–æ—Å—Ç–æ –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º "no_rating".

```python
games['rating'] = games['rating'].fillna(value='no_rating') 
```

```python
games['rating'].isna().sum()
```

–¢.–∫. –Ω–∞ –ø—Ä–æ–¥–∞–∂—É –∫–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã—Ö –∏–≥—Ä –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–æ –≤–ª–∏—è—é—Ç –∫–æ–Ω—Å–æ–ª–∏, —Ç–æ –∑–∞ –æ—Ç—Å—á–µ—Ç –º–æ–∂–Ω–æ –≤–∑—è—Ç—å –≤—ã—Ö–æ–¥ –º–∞—Å—Å–æ–≤—ã—Ö –∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∫–æ–Ω—Å–æ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –≤–ª–∏—è—é—Ç –Ω–∞ –∏–Ω–¥—É—Å—Ç—Ä–∏—é –¥–æ —Å–∏—Ö –ø–æ—Ä: playstation 3 (2006), xbox 360 (2005), nintendo ds (2004), –∏–ª–∏ —Å–∞–º–∞—è –ø—Ä–æ–¥–∞–≤–∞–µ–º–∞—è –∫–æ–Ω—Å–æ–ª—å –≤ –º–∏—Ä–µ playstation 2 (2000). 

### –ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

–í–∑–≥–ª—è–Ω–µ–º –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.

```python
pd.pivot_table(games,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index=['platform'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values=['sales'],
               #–¥–∞–Ω–Ω—ã–µ —Å—É–º–º–∏—Ä—É–µ–º
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, —Å–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫
               fill_value=0).sort_values(by=['sales'], ascending=False).plot.bar(figsize=(11, 5), color='dodgerblue', width=0.6)
plt.grid()
plt.title("–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º")
plt.xlabel("–ü–ª–∞—Ç—Ñ–æ—Ä–º—ã")
plt.ylabel("–º–ª–Ω. –∫–æ–ø–∏–π");
```

```python
#—Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ games
top_5_platforms = pd.pivot_table(games,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index=['platform'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values=['sales'],
               #–¥–∞–Ω–Ω—ã–µ —Å—É–º–º–∏—Ä—É–µ–º
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, —Å–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ —É–±—ã–≤–∞–Ω–∏—é –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫
               fill_value=0).sort_values(by=['sales'], ascending=False).head()

top_5_platforms
```

–ú—ã –≤–∏–¥–∏–º, —á—Ç–æ —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —ç—Ç–æ PS2, X360, PS3, Wii –∏ DS.

```python
platforms_sales = pd.pivot_table(games,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –≥–æ–¥ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä—ã
               index=['year_of_release'],
               #–∑–∞ –∫–æ–ª–æ–Ω–∫–∏ - –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º
               columns=['platform'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values=['sales'],
               #–¥–∞–Ω–Ω—ã–µ —Å—É–º–º–∏—Ä—É–µ–º
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0
               fill_value=0)
#—É–¥–∞–ª—è–µ–º –≤–µ—Ä—Ö–Ω–µ–µ –Ω–∞—Ö–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤
platforms_sales = platforms_sales.rename(columns={"sales":""})
platforms_sales.columns = [t[0] if t[0] else t[1] for t in platforms_sales.columns]
#–æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏–∑ –Ω–∞—à–µ–≥–æ —Ç–æ–ø5
platforms_sales = platforms_sales[["PS2", "X360", "PS3", "Wii", "DS"]]
#–≤—ã–≤–æ–¥–∏–º –≥—Ä–∞—Ñ–∏–∫
platforms_sales.plot(figsize=(15, 5))
plt.title('–ü—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º')
plt.xlabel('–ì–æ–¥')
plt.ylabel('–º–ª–Ω. –∫–æ–ø–∏–π')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
games_filtered = games.query('year_of_release >= 2012')
```

### –ö–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ª–∏–¥–∏—Ä—É—é—Ç –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º, —Ä–∞—Å—Ç—É—Ç –∏–ª–∏ –ø–∞–¥–∞—é—Ç?

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ –∫–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å–µ–π—á–∞—Å –ª–∏–¥–∏—Ä—É—é—Ç, —Å–¥–µ–ª–∞–µ–º —Å–±–æ—Ä–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –≥–¥–µ —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ –≥–æ–¥–∞–º.

```python
actual_games = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –≥–æ–¥ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä—ã
               index=['year_of_release'],
               #–∑–∞ –∫–æ–ª–æ–Ω–∫–∏ - –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º
               columns=['platform'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values=['sales'],
               #–¥–∞–Ω–Ω—ã–µ —Å—É–º–º–∏—Ä—É–µ–º
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0
               fill_value=0)
#—É–¥–∞–ª—è–µ–º –≤–µ—Ä—Ö–Ω–µ–µ –Ω–∞—Ö–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤
actual_games = actual_games.rename(columns={"sales":""})
actual_games.columns = [t[0] if t[0] else t[1] for t in actual_games.columns]
#–í—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
actual_games

```

```python
#–≤—ã–≤–æ–¥–∏–º –≥—Ä–∞—Ñ–∏–∫
actual_games.plot(figsize=(14, 7))
plt.title('–ü—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –∑–∞ 5 –ª–µ—Ç')
plt.xlabel('–ì–æ–¥')
plt.ylabel('–º–ª–Ω. –∫–æ–ø–∏–π')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

–ò –∏–∑ –≥—Ä–∞—Ñ–∏–∫–∞, –∏ –∏–∑ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä, —á–µ–º –≤—ã—à–µ –æ—Ü–µ–Ω–∫–∞, —Ç–µ–º –±–æ–ª—å—à–µ –ø—Ä–æ–¥–∞–µ—Ç—Å—è –∏–≥—Ä–∞.

–ú—ã –≤–∏–¥–∏–º, —á—Ç–æ –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø—Ä–æ—Å–ª–µ–∂–∏–≤–∞–µ—Ç—Å—è –Ω–µ —Ç–∞–∫–∞—è –∂–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å, –∫–∞–∫ —É X360. –ú–æ–∂–µ—Ç –±—ã—Ç—å –¥–µ–ª–æ –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ —Ç–∞–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫, –ø—Ä–æ–≤–µ—Ä–∏–º.

```python

pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index=['platform'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
               values=['user_score'],
               #—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ü–µ–Ω–æ–∫
               aggfunc='count',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å
               fill_value=0)
```

–ú—ã –≤–∏–¥–∏–º, —á—Ç–æ –æ—Ü–µ–Ω–æ–∫ –≤ –Ω–∞—à–µ–º –≤—Ä–µ–º–µ–Ω–Ω–æ–º –ø–µ—Ä–∏–æ–¥–µ –¥–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º DS, PSP, Wii –æ—á–µ–Ω—å –º–∞–ª–æ, –ø–æ –Ω–∏–º –¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã –Ω–µ–ª—å–∑—è.
–î–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –Ω–µ–ª—å–∑—è —Å–¥–µ–ª–∞—Ç—å –æ–¥–Ω–æ–∑–Ω–∞—á–Ω—ã–π –≤—ã–≤–æ–¥, —á—Ç–æ –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏, –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º WiiU, –≥–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–∞–≤–µ–Ω 0,4, –∏ –º—ã –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –¥–ª—è —ç—Ç–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –µ—Å—Ç—å –ø—Ä—è–º–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏.


–û—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤:

–ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ —Å –±–æ–ª—å—à–∏–º –æ—Ç—Ä—ã–≤–æ–º –∏–≥—Ä –∂–∞–Ω—Ä–∞ Action, –Ω–∞ –≤—Ç–æ—Ä–æ–º –º–µ—Å—Ç–µ Role-Playing, –∞ –Ω–∞ —Ç—Ä–µ—Ç—å–µ–º Adventure. –ú–µ–Ω—å—à–µ –≤—Å–µ–≥–æ –∏–≥—Ä –∂–∞–Ω—Ä–∞ Puzzle.

```python
pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index=['genre'],
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values=['sales'],
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).sort_values(by='sales', ascending=False)
```

–ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø—Ä–æ–¥–∞–∂ —Ç–æ–∂–µ —É –∂–∞–Ω—Ä–∞ Action, –≤ –º–µ–Ω—å—à–µ —É Puzzle. –≠—Ç–æ –ª–æ–≥–∏—á–Ω–æ, –≤–µ–¥—å —á–µ–º –±–æ–ª—å—à–µ –∏–≥—Ä –Ω–∞ —Ä—ã–Ω–∫–µ, —Ç–µ–º –±–æ–ª—å—à–µ –∏—Ö –º–æ–≥—É—Ç –ø–æ–∫—É–ø–∞—Ç—å. –ü–æ—Å—á–∏—Ç–∞–µ–º –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏–≥—Ä –∂–∞–Ω—Ä–∞ –Ω–∞ –∏—Ö –ø—Ä–æ–¥–∞–∂–∏.

 - –û—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–µ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä, –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º WiiU, –≥–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Ä–∞–≤–µ–Ω 0,4, –∏ –º—ã –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –¥–ª—è —ç—Ç–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –µ—Å—Ç—å –ø—Ä—è–º–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É –æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏.

## –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

### –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏–∑ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–∏ (—Ä–µ–≥–∏–æ–Ω NA)

#### –¢–æ–ø 5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º

–û–ø—Ä–µ–¥–µ–ª–∏–º —Ç–æ–ø 5 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–∞–∂–∞—Ö –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ

```python
na_platform_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='platform',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='na_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['na_sales'], ascending=False).head()
na_platform_sales
```

–î–æ–±–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–µ–º others

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_na_sales = games_filtered['na_sales'].sum() - na_platform_sales['na_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'platform': 'others', 'na_sales': others_na_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
na_platform_sales = na_platform_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
na_platform_sales = na_platform_sales.set_index('platform')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
na_platform_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

```python
na_platform_sales['na_sales'].plot(kind="pie",
                                           autopct='%1.1f%%',
                                           explode=[0.05, 0.01, 0.01, 0.01, 0.01, 0.01],
                                           legend=True,
                                           title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ',
                                           ylabel='',
                                           labeldistance=None,
                                           figsize=(6, 6),
                                           cmap='rainbow')
plt.legend()
plt.show()
```

–ü—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ X360 —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 23,7%, PS4 - 18,4%, PS3 - 17,5%. –≠—Ç–∏ —Ç—Ä–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∑–∞–Ω–∏–º–∞—é—Ç –±–æ–ª—å—à–µ –ø–æ–ª–æ–≤–∏–Ω—ã —Ä—ã–Ω–∫–∞.

#### –¢–æ–ø 5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤

–û–ø—Ä–µ–¥–µ–ª–∏–º —Ç–æ–ø 5 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–∞–∂–∞—Ö –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ

```python
na_genre_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='genre',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='na_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['na_sales'], ascending=False).head()
na_genre_sales
```

–î–æ–±–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–µ–º others

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –∂–∞–Ω—Ä—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_na_sales = games_filtered['na_sales'].sum() - na_genre_sales['na_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∂–∞–Ω—Ä–æ–≤ –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'genre': 'others', 'na_sales': others_na_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
na_genre_sales = na_genre_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
na_genre_sales = na_genre_sales.set_index('genre')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
na_genre_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

```python
na_genre_sales['na_sales'].plot(kind="pie",
                                           autopct='%1.1f%%',
                                           explode=[0.05, 0.01, 0.01, 0.01, 0.01, 0.01],
                                           legend=True,
                                           title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –∂–∞–Ω—Ä–æ–≤ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ',
                                           ylabel='',
                                           labeldistance=None,
                                           figsize=(6, 6),
                                           cmap='rainbow')
plt.legend()
plt.show()
```

–ü—Ä–æ–¥–∞–∂–∏ –∂–∞–Ω—Ä–∞ Action —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 30,1%, Shooter - 24,5%, Sports - 13,8%. –≠—Ç–∏ —Ç—Ä–∏ –∂–∞–Ω—Ä–∞ –∑–∞–Ω–∏–º–∞—é—Ç –±–æ–ª—å—à–µ –¥–≤—É—Ö —Ç—Ä–µ—Ç–µ–π —Ä—ã–Ω–∫–∞.

#### –í–ª–∏—è–µ—Ç –ª–∏ —Ä–µ–π—Ç–∏–Ω–≥ ESRB –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ

–û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–µ —Ä–µ–π—Ç–∏–Ω–≥–∏, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–∞–∂–∞—Ö –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ

```python
na_rating_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='rating',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='na_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).sort_values(by=['na_sales'], ascending=False).head()
na_rating_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

```python
na_rating_sales['na_sales'].plot(kind="pie",
                                           autopct='%1.1f%%',
                                           explode=[0.05, 0.01, 0.01, 0.01, 0.01],
                                           legend=True,
                                           title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ',
                                           ylabel='',
                                           labeldistance=None,
                                           figsize=(6, 6),
                                           cmap='rainbow')
plt.legend()
plt.show()
```

–ù–∞–ø–æ–º–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤:

–ì—Ä–∞—Ñ–∏–∫–∏ –≤–º–µ—Å—Ç–µ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏.

```python
#–∑–∞–¥–∞–µ–º —Ä–∞–∑–º–µ—Ä—ã –æ–±—â–µ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
fig, ax = plt.subplots(1, 3, figsize=(15,12))
fig.subplots_adjust(top=1.4, wspace= 0.5,)
fig.suptitle('–°–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞', fontsize=20)
#—Å–æ–∑–¥–∞–µ–º –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –ø–æ –∂–∞–Ω—Ä–∞–º
ax[0].pie(na_platform_sales['na_sales'],
          labels=na_platform_sales.index,
          radius = 1.1,
          autopct='%.0f%%')
ax[0].set_title('–°–∞–º—ã–π –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã', fontsize=12)


# —Å–æ–∑–¥–∞–µ–º –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –∂–∞–Ω—Ä–∞–º:
ax[1].pie(na_genre_sales['na_sales'],
          labels=na_genre_sales.index,
          radius = 1.1,
          autopct='%.0f%%')
ax[1].set_title('–°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã', fontsize=12)

# —Å–æ–∑–¥–∞–¥–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Ä–µ–π—Ç–∏–Ω–≥—É:
ax[2].pie(na_rating_sales['na_sales'],
          labels=na_rating_sales.index,
          radius = 1.1,
          autopct='%.0f%%')
ax[2].set_title('–°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ —Ä–µ–π—Ç–∏–Ω–≥–∏', fontsize=12)

plt.show()
```

 - –∏–≥—Ä–∞–µ—Ç –Ω–∞ –∫–æ–Ω—Å–æ–ª—è—Ö, –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞–µ—Ç –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã X360, PS4, PS3,
 - –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞–µ—Ç –∏–≥—Ä—ã –≤ –∂–∞–Ω—Ä–∞—Ö Action, Shooter, Sports,

### –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏–∑  –ï–≤—Ä–æ–ø—ã (—Ä–µ–≥–∏–æ–Ω EU)

#### –¢–æ–ø 5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º

–û–ø—Ä–µ–¥–µ–ª–∏–º —Ç–æ–ø 5 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ –ø—Ä–æ–¥–∞–∂–∞—Ö –≤ –ï–≤—Ä–æ–ø–µ.

```python
eu_platform_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='platform',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='eu_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['eu_sales'], ascending=False).head()
eu_platform_sales
```

–î–æ–±–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–µ–º others

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_eu_sales = games_filtered['eu_sales'].sum() - eu_platform_sales['eu_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'platform': 'others', 'eu_sales': others_eu_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
eu_platform_sales = eu_platform_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
eu_platform_sales = eu_platform_sales.set_index('platform')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
eu_platform_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

```python
eu_platform_sales['eu_sales'].plot(kind="pie",
                                           autopct='%1.1f%%',
                                           explode=[0.05, 0.01, 0.01, 0.01, 0.01, 0.01],
                                           legend=True,
                                           title='–ü–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤ –ï–≤—Ä–æ–ø–µ',
                                           ylabel='',
                                           labeldistance=None,
                                           figsize=(6, 6),
                                           cmap='rainbow')
plt.legend()
plt.show()
```

–ü—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ PS4 —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 27,8%, PS3 - 21,1%, X360 - 14,7%. –≠—Ç–∏ —Ç—Ä–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∑–∞–Ω–∏–º–∞—é—Ç –±–æ–ª—å—à–µ –ø–æ–ª–æ–≤–∏–Ω—ã —Ä—ã–Ω–∫–∞.

#### –¢–æ–ø 5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –∂–∞–Ω—Ä–æ–≤

```python
eu_genre_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='genre',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='eu_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['eu_sales'], ascending=False).head()
eu_genre_sales
```

–î–æ–±–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–µ–º others

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –∂–∞–Ω—Ä—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_eu_sales = games_filtered['eu_sales'].sum() - eu_genre_sales['eu_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∂–∞–Ω—Ä–æ–≤ –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'genre': 'others', 'eu_sales': others_eu_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
eu_genre_sales = eu_genre_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
eu_genre_sales = eu_genre_sales.set_index('genre')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
eu_genre_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

–ü—Ä–æ–¥–∞–∂–∏ –∂–∞–Ω—Ä–∞ Action —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç 31,4%, Shooter - 22,4%, Sports - 13,6%. –≠—Ç–∏ —Ç—Ä–∏ –∂–∞–Ω—Ä–∞ –∑–∞–Ω–∏–º–∞—é—Ç –±–æ–ª—å—à–µ –¥–≤—É—Ö —Ç—Ä–µ—Ç–µ–π —Ä—ã–Ω–∫–∞.

### –ü–æ—Ä—Ç—Ä–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏–∑  –Ø–ø–æ–Ω–∏–∏ (—Ä–µ–≥–∏–æ–Ω JP)

#### –¢–æ–ø 5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º

```python
jp_platform_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='platform',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='jp_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['jp_sales'], ascending=False).head()
jp_platform_sales
```

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_jp_sales = games_filtered['jp_sales'].sum() - jp_platform_sales['jp_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'platform': 'others', 'jp_sales': others_jp_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
jp_platform_sales = jp_platform_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
jp_platform_sales = jp_platform_sales.set_index('platform')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
jp_platform_sales
```

–ü–æ–∫–∞–∂–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã.

```python
jp_genre_sales = pd.pivot_table(games_filtered,
               #–∑–∞ –∏–Ω–¥–µ–∫—Å –≤–æ–∑—å–º–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
               index='genre',
               #–î–∞–Ω–Ω—ã–µ –±–µ—Ä–µ–º –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ø—Ä–æ–¥–∞–∂
               values='jp_sales',
               #—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É
               aggfunc='sum',
               #–ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–ø–æ–ª–Ω—è–µ–º 0, –µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
               fill_value=0).reset_index().sort_values(by=['jp_sales'], ascending=False).head()
jp_genre_sales
```

```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—É–º–º—É –ø—Ä–æ–¥–∞–∂ –Ω–∞ –∂–∞–Ω—Ä—ã, –Ω–µ –≤–æ—à–µ–¥—à–∏–µ –≤ —Ç–æ–ø 5
others_jp_sales = games_filtered['jp_sales'].sum() - jp_genre_sales['jp_sales'].sum()
#—Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∂–∞–Ω—Ä–æ–≤ –∏ —Å—É–º–º–æ–π –∏—Ö –ø—Ä–æ–¥–∞–∂
new_row = {'genre': 'others', 'jp_sales': others_jp_sales}
#–¥–æ–±–∞–≤–∏–º –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É –∫ –Ω–∞—à–µ–º—É –¥–∞—Ç–∞—Ñ—Ä–µ–º—É
jp_genre_sales = jp_genre_sales.append(new_row, ignore_index = True )
#—É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—É, –∫–∞–∫ –∏–Ω–¥–µ–∫—Å
jp_genre_sales = jp_genre_sales.set_index('genre')
#–≤—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É
jp_genre_sales
```

## –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

### –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ

–í—ã–¥–≤–µ–Ω–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC —Ä–∞–≤–Ω—ã. –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ - —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –Ω–µ —Ä–∞–≤–Ω—ã.  
–ü—Ä–æ–≤–µ—Ä–∏–º –µ–µ –º–µ—Ç–æ–¥–æ–º –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –∏–∑ –Ω–∏—Ö –≤—ã–±–æ—Ä–∫–∞–º

```python
# –∑–∞–¥–∞—ë–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
alpha = 0.05
# –ø—Ä–æ–≤–æ–¥–∏–º ttest –¥–ª—è –¥–≤—É—Ö –≤—ã–±–æ—Ä–æ–∫, –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π –∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–≤—É—Ö—Å—Ç–æ—Ä–æ–Ω–Ω—é—é –≥–∏–ø–æ—Ç–µ–∑—É 
result = st.ttest_ind(games_filtered.query('platform == "XOne" & user_score >= 0')['user_score'],
                      games_filtered.query('platform == "PC" & user_score >= 0')['user_score'],
                      equal_var=False,
                      alternative='two-sided')

# –ø—Ä–æ–≤–µ—Ä—è–µ–º p-value
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', result.pvalue)

if result.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É') 
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã Xbox One:', 
      games_filtered.query('platform == "XOne"')['user_score'].mean())
print('–°—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã PC:', 
      games_filtered.query('platform == "PC"')['user_score'].mean())
```
